```
# Getting needed libraries
import matplotlib.pyplot as plt
import math
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

## 1.1 Create a two-dimensional vector and plot it on a graph

```
# creating a function that plot 2-dim vector
def graph_2dim_vectors(v):
  plt.arrow(0,0,v[0],v[1], width=0.5)
  plt.xlim(-1,9); plt.ylim(-1,9)
  return plt.show()
```

```
vector=[2.1,3.5]
graph_2dim_vectors(vector)
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
# creating a function that plot 3-dim vector
def graph_3dim_vectors(v):
  v_v_3d=np.array([[0,0,0,v[0],v[1],v[2]]])
  x,y,z,u,v,w=zip((*v_v_3d))
  fig=plt.figure()
  ax=fig.add_subplot(111,projection="3d")
  ax.quiver(x,y,z,u,v,w,length=1)
  ax.set_xlim([0,4])
  ax.set_ylim([0,4])
  ax.set_zlim([0,4])
  ax.set_xlabel("X")
  ax.set_ylabel("Y")
  ax.set_zlabel("Z")
  plt.show()
```

```
vector=[1.6,3.8,2.3]
graph_3dim_vectors(vector)
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
# creating a scalling function
def scale_vector(v,x):
  result=np.multiply(v,x)
  return list(result)
```

```
# scalling the vector with different numbers
vector=[1,0.9]
vector_5=scale_vector(vector,5)
vector_pi=scale_vector(vector,math.pi)
vector_e=scale_vector(vector,math.e)
```

```
# graphing4 vectors  in one plane
V = np.array([vector, vector_5, vector_pi])
origin = np.array([[0, 0, 0],[0, 0, 0]]) # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=21)

plt.quiver(*origin, vector_e[0], vector_e[1])


plt.show()
```

## 1.6 Find $c \cdot d$

\vec{c} = \begin{bmatrix}7 & 22 & 4 & 16\end{bmatrix}

```
c=[7,22,4,16]
d=[12,6,2,9]
# let a = c .d which means the dot product between c and d
# Note: the dot product is always a number
a=np.dot(c,d)
a
```

```
# here is anothe way to compute the dot product between c and d
sum=0
for i in range(4):
  sum+=c[i]*d[i]
print(sum)
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
#finding the magnitude of g
g=[1,1,1,8]
magnitude =  np.sqrt(np.dot(g,g))
magnitude
```

```
# Another way to do it
sum=0
for item in g:
  sum+=item**2
print(math.sqrt(sum))
```

```
#finding the magnitude of h
h=[3,3,3,3]
magnitude =  np.sqrt(np.dot(h,h))
magnitude
```
